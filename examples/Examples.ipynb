### Comprehensive Usage

```python
# Import toy dataset (raw dataset)
df = load_data('Fish')
df.head()
```

```python
assume = Check(df=df, 
               target='Weight',
               task='linear regression',
               predictors=['Height', 'Width', 'Length1', 'Species'],
               keep=True,
               categorical_features=['Species'],
               categorical_encoder='ohe',
               mode='inline')

assume.report()
```
